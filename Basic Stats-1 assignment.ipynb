### normalization of cateogrical values using one hot encoding method

```python
from sklearn.preprocessing import OneHotEncoder,LabelEncoder
```

```python
ohencoder=OneHotEncoder()
label=LabelEncoder()
```

when doing onehot encoder for ordinal values it gives us increse in no of features or gives a complex data set

```python
##for example
df['SKU'].unique()
```

there are lot of 'SKU' values so if we apply onehot encoding to this it will result in a lot of  binary equivalent columns
so we an apply here label encoder

```python
df['SKU']=label.fit_transform(df[['SKU']])
```

```python
df['Day'].unique()
```

same case in DAY column so apply label encoding

```python
df['Day']=label.fit_transform(df[['Day']])
```

```python
df['City'].unique()
```

there is only one city so here we can use one-hot encoding

```python
onearray=ohencoder.fit_transform(df[['City']])
```

```python
df['City']=pd.DataFrame(onearray.toarray())
```

```python
df['City']
```

```python
df['BU'].unique()
```

apply label encoder to 'BU'

```python
df['BU']=label.fit_transform(df[['BU']])
```

```python
df['Brand'].unique()
```

```python
df['Brand']=label.fit_transform(df[['Brand']])
```

```python
df['Model'].unique()
```

```python
df['Model']=label.fit_transform(df[['Model']])
```
